```python

df1=df.groupby("Continent")[["2022 Population"]].sum()
df1
```

```python
import seaborn as sns
sns.barplot(x=df1.index,y=df1["2022 Population"],order=["Asia","Africa","Europe","North America","South America",
                                                      "Oceania" ],color="red")
```
